```python
df_wine.drop( columns=['density','pH'], inplace=True )
```

```python
df_wine
```

```python
bins = (2, 6.5, 8)
group_names= ['bad','good']
df_wine['quality'] = pd.cut( df_wine['quality'], bins= bins, labels = group_names  )
```

```python
df_wine.sort_values( by='quality', ascending=False )
```

```python
label_quality = LabelEncoder()

df_wine['quality'] = label_quality.fit_transform( df_wine['quality'] )

```

```python
label_quality.classes_
```

```python
df_wine['quality'].value_counts()
```

```python
sns.countplot(df_wine['quality'])
```

```python
X = df_wine.drop('quality', axis=1 )
y= df_wine['quality']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2 )
```

```python
X_train
```

```python
X_train.describe()
```

```python
sc = StandardScaler()
```

```python
X_train = sc.fit_transform(X_train)
X_test =  sc.fit_transform(X_test)
```

```python
pd.DataFrame(X_train).describe()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier( n_estimators=200 )
rfc.fit(X_train, y_train)
pred_rfc = rfc.predict( X_test )
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, pred_rfc))
```
