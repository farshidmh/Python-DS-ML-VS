```python
def categorize_sales(price):
  if price < 50:
    return "Low"
  elif price < 200:
    return "Medium"
  else:
    return "High"
```

```python
ds =  df[ ['Customer Name','State','Sales'] ].sample(100)
```

```python
ds
```

```python
ds['Sales Price Category']  = ds['Sales'].apply(categorize_sales)
```

```python
ds
```

```python
ds['Customer Name Length'] = ds['Customer Name'].apply(len)
```

```python
ds['Discounted Price'] = ds['Sales'].apply(  lambda x: x*0.85 if x > 200 else x  )
```
