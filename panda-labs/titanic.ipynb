```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()

df_titanic['Sex'] = le.fit_transform( df_titanic['Sex'] )
df_titanic['Embarked'] = le.fit_transform( df_titanic['Embarked'] )
```

```python
le.classes_
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
X = df_titanic.drop( columns=['Survived'] )
y = df_titanic['Survived']
```

```python
scaler = StandardScaler()

X = scaler.fit_transform(X)
X = pd.DataFrame(X)
```

```python
X
```

```python
y
```

```python
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, t_test = train_test_split( X, y, test_size=0.3 )
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
model_lr = LogisticRegression()
model_lr.fit(X_train, y_train)
```

```python
model_dt = DecisionTreeClassifier()
model_dt.fit(X_train, y_train)
```

```python
model_rf = RandomForestClassifier()
model_rf.fit(X_train, y_train)
```

```python
model_kn = KNeighborsClassifier()
model_kn.fit(X_train, y_train)
```

```python
y_pred_lr = model_lr.predict( X_test )
y_model_dt = model_dt.predict( X_test )
y_model_rf = model_rf.predict( X_test )
y_model_kn = model_kn.predict( X_test )
```

```python
accuracy_score(t_test, y_pred_lr)
```

```python
accuracy_score(t_test, y_model_dt)
```

```python
accuracy_score(t_test, y_model_rf)
```

```python
accuracy_score(t_test, y_model_kn)
```
