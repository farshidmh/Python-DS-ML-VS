```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
covid = pd.read_csv('https://raw.githubusercontent.com/fenago/data-analysis-pandas/master/lab_05/data/covid19_cases.csv').assign(
    date = lambda x: pd.to_datetime(  x.dateRep , format='%d/%m/%Y'  )
).set_index('date').replace('United_States_of_America','USA').sort_index()['2020-01-01':'2020-10-31']
```

```python
covid
```

```python
ncr = covid.pivot_table(
    index=covid.index,
    columns='countriesAndTerritories',
    values='cases'
).rolling(7).mean()
ncr
```

```python
fig, axes = plt.subplots(1,3, figsize=(20,7))

ncr[ ['China'] ].plot( ax=axes[0] , style='-.c' )
ncr[ ['Italy','Spain']  ].plot(
    ax=axes[1], style=['-','--'],
    title='7-day rolling avg of new cases '
)
ncr[ ['Brazil','India','USA']  ].plot(
    ax=axes[2], style=['-','--',':'],
)


plt.show()
```
